```python
Cl = '/content/Classeurbb.xlsx'
data= pd.read_excel(Cl,
sheet_name=0,
header=0,
index_col=False,
keep_default_na=True)
```

```python
del data["Source"]
del data["CodeSite"]
```

```python
data
```

```python
X1 = data.loc[:, data.columns != 'Cx_theileri']
y1 = data.Cx_theileri
```

```python
X1
```

```python
# Train/test split
X_train, X_test, y_train, y_test = train_test_split(
X1,y1,test_size=0.20)
```

```python
print('Training dataset shape:', X_train.shape, y_train.shape)
print('Testing dataset shape:', X_test.shape, y_test.shape)
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector as sfs
# Build RF classifier to use in feature selection
clf =  RandomForestClassifier(n_estimators=100, criterion='gini', max_depth=7, random_state=11)

# Build step forward feature selection
sfsT = sfs(clf,
           k_features=20,
           forward=True,
           floating=False,
           verbose=2,
           scoring='roc_auc',
           cv=5)

# Perform SFFS
sfsT = sfsT.fit(X_train, y_train)
```

```python
feat_cols = list(sfsT.k_feature_idx_)
print(feat_cols)
```
