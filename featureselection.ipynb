```python
Cl = '/content/cs-longiaN+.xlsx'
data= pd.read_excel(Cl,
sheet_name=0,
header=0,
index_col=False,
keep_default_na=True)
```

```python
del data["Source"]
del data["CodeSite"]
```

```python
data
```

```python
X1 = data.loc[:, data.columns != 'Cs_longia']
y1 = data.Cs_longia
```

```python
# Train/test split
X_train, X_test, y_train, y_test = train_test_split(
X1,y1,test_size=0.20)
```

```python
X_train
```

```python
y_train
```

```python
print('Training dataset shape:', X_train.shape, y_train.shape)
print('Testing dataset shape:', X_test.shape, y_test.shape)
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector as sfs
# Build RF classifier to use in feature selection
clf = XGBClassifier(scale_pos_weight=1.5,learning_rate=0.1,max_depth=5,n_estimators=100)

# Build step forward feature selection
sfsT = sfs(clf,
           k_features=20,
           forward=True,
           floating=False,
           verbose=2,
           scoring='roc_auc',
           cv=5)

# Perform SFFS
sfsT = sfsT.fit(X_train, y_train)
```

```python
feat_cols = list(sfsT.k_feature_idx_)
print(feat_cols)
```

```python
data.columns[feat_cols]
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector as sfs
# Build RF classifier to use in feature selection
clf = XGBClassifier(scale_pos_weight=1.5,learning_rate=0.1,max_depth=5,n_estimators=100)

# Build step forward feature selection
sfs2 = sfs(clf,
           k_features=11,
           forward=True,
           floating=False,
           verbose=2,
           scoring='roc_auc',
           cv=5)

# Perform SFFS
sfs2 = sfs2.fit(X_train, y_train)
```

```python
feat_cols2 = list(sfs2.k_feature_idx_)
print(feat_cols2)
```

```python
feature=data.columns[feat_cols2]
```

```python
feature
```

```python
data[feature]
```

```python
T_train=X_train[feature]
```

```python
T_train
```

```python
T_Test=X_test[feature]
```

```python
T_Test
```

```python
from sklearn.metrics import accuracy_score as acc
# Build full model with selected features
clf = XGBClassifier(scale_pos_weight=1.5,learning_rate=0.1,max_depth=5,n_estimators=100)
clf.fit(T_train, y_train)

y_train_pred = clf.predict(T_train)
print('Training accuracy on selected features: %.3f' % acc(y_train, y_train_pred))

y_test_pred = clf.predict(T_Test)
print('Testing accuracy on selected features: %.3f' % acc(y_test, y_test_pred))
```

```python
from sklearn.metrics import matthews_corrcoef
def tp1(y_true, y_pred): 
    MCC= matthews_corrcoef(y_true, y_pred)
    return MCC
```

```python
tp1(y_test,y_test_pred)
```
